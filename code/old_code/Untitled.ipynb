```python
import pandas as pd
import numpy as np
import os
import nibabel
```

```python
# Find all files in the structural data folder
data_path = './../data/MoAEpilot/sM00223/'
files = os.listdir(data_path)
```

```python
import matplotlib.pyplot as plt
import nibabel

data = nibabel.load('./../data/MoAEpilot/sM00223/sM00223_002.hdr').get_fdata()
```

```python
print(data.shape)
```

```python
fig, ax = plt.subplots(1, 8, figsize=[18, 3])
n = 0
for slice in range(10,18):
    ax[n].imshow(data[:, :, slice], 'viridis')
    ax[n].set_xticks([])
    ax[n].set_yticks([])
    ax[n].set_title('Slice number: {}'.format(slice), color='r')
    n += 1  
fig.subplots_adjust(wspace=0, hspace=0)
plt.show()
```
